# Problem 038:  Pandigital Multiples

> Take the number $192$ and multiply it by each of $1$, $2$, and $3$:
> 
> $$\begin{align}
192 \times 1 &= 192\\
192 \times 2 &= 384\\
192 \times 3 &= 576
\end{align}$$
>
> By concatenating each product we get the $1$ to $9$ pandigital, $192384576$. We will call $192384576$ the concatenated product of $192$ and $(1,2,3)$.
>
> The same can be achieved by starting with $9$ and multiplying by $1$, $2$, $3$, $4$, and $5$, giving the pandigital, $918273645$, which is the concatenated product of $9$ and $(1, 2, 3, 4, 5)$.
>
> What is the largest $1$ to $9$ pandigital $9$-digit number that can be formed as the concatenated product of an integer with $(1,2, \dots, n)$ where $n \gt 1$?

## Solution $\mathcal O\left(M 10^{\lfloor M / 2\rfloor}\right)$

Fairly straightforward.  I guess the search _could_ be done in two different ways: by searching for root of the concatenating product or by looking if a permutation can be classified as a concatenated producted.  For a list of $M$ unique digits that you want the concatenated product to contain, there are $10^{\lfloor M / 2 \rfloor}$ values to look through for the former, and $M!$ values for the latter, making the first more efficient.  Also, it would be non-trivial to determine if a given permutation would be a concatenated product of some value.

I loop through all possible "roots" of the concatenated product, like the value $192$ in the problem statement.  I make a count of all the digits that need to be in the answer, and then mark them off one by one with the digits of the root.  Then I double the root and repeat.  Either a digit that has been used will be seen again and the process is stopped, or all $9$ digits will have been seen and the value is stored.  The root must be less than or equal to the number of digits in the concatenated product, since at least two values need to be strung together.  I make the code a little more general by allowing a target list of digits with any count, not just a single count of the digits $1-9$.  With $10^{\lfloor M / 2\rfloor}$ values to go through, and having to look over $M$ digits for each one, the time scaling is $\mathcal O\left(10^{\lfloor M / 2\rfloor}\right)$.  Some time could be saved by ordering which integers to look at first, for example the problem itself identifies that there is a concatenating product that starts with $9$, thus the root must also start with $9$, but this is fast enough.

```python
%%time

def get_digits(x: int) -> int:
    while x:
        yield x % 10
        x //= 10

def p038(digits0: list[int]) -> str:
    val = "0"

    nd0 = sum(digits0)
    
    imax = pow(10, nd0 // 2)
    
    for i in range(1, imax):
        digits = digits0.copy()
        nd = 0
        igo = True
        j = 1
        nstr = ""
        while igo:
            for d in get_digits(i*j):
                if not digits[d]:
                    igo = False
                    break
                digits[d] -= 1
                nd += 1
            else:
                nstr += str(i*j)
                if nd == nd0:
                    igo = False
                    val = max(val, nstr)
            j += 1
        
    return val

digits0 = [0] + [1] * 9
ans = p038(digits0)

print(ans)
```
