# Problem 037:  Truncatable Primes

> The number $3797$ has an interesting property. Being prime itself, it is possible to continuously remove digits from left to right, and remain prime at each stage:  $3797$, $797$, $97$, and $7$. Similarly we can work from right to left: $3797$, $379$, $37$, and $3$.
>
> Find the sum of the only eleven primes that are both truncatable from left to right and right to left.
>
> NOTE:  $2$, $3$, $5$, and $7$ are not considered to be truncatable primes.

## Solution

I usually like using a prime sieve any time that prime numbers are involved, but there are a few aspects of this problem that make it not in our interest to apply one here.  First, there are relatively not that many truncatable primes, so finding all primes and searching through them is inefficient as opposed to a dynamic programming approach.  There is also the problem of not knowing the maximum prime that we will have to find, even if my previously used extendable prime algorithm can be used.  So, with only needing to check a handful of numbers, it is efficient to just do trial division from $2$ to $\lfloor \sqrt n \rfloor$ for the number $n$.

With such an open ended problem with no immediate sense of what the largest number will be, some assumptions or outside information is useful to limit the search.  The problem does tell us that there are a limited number of these primes, and that there are exactly 11.  The assumption I use in the below code is that there are a finite number of right truncatable primes.  I cannot think of a succinct proof that this would be true, but it is.  References like the Wikipeida page for [Truncatable primes](https://en.wikipedia.org/wiki/Truncatable_prime) can provide this insight, but also gives a means of finding the answer with no code necessary.  I can show that the number of $1$'s and $7$'s are limited in the prime, but the $3$'s and $9$'s are harder.

So, that said, this is my approach.  I use dynamic programming to grow the right truncatable prime starting with the most significan figure.  The first number must be a single digit prime, so $2$, $3$, $5$, or $7$.  All following digits cannot be even or $5$, since the resulting value will be divisible by $2$ or $5$, respectively.  That leave $1$, $3$, $7$, and $9$.  Every time I find a prime value with this scheme I then check if the number is left truncatable, and add it if so.  The assumption here is that the code will terminate and the size of right truncatable primes are small enough to efficiently find primality.  As another example, consider doing the problem in the other direction, find all the left truncatable primes and then determine if they are right truncatable.  Even though there are a finite number of these values, this code would lead to finding the primality of $24$ digit numbers, which would be infeasible.

```python
%%time

def is_prime(x: int) -> bool:
    if x < 2:
        return False
    imax = int(x**0.5) + 1 
    return all(x % i for i in range(2, imax))

def p037() -> int:
    val = 0

    first = [2, 3, 5, 7]
    middle = [1, 3, 7, 9]
    digits = [0, 0]
    while digits:
        n = first[digits[0]]
        for d in digits[1:]:
            n = 10 * n + middle[d]

        if is_prime(n):
            pow10 = 1
            while (pow10 := 10 * pow10) < n:
                if not is_prime(n % pow10):
                    break
            else:
                val += n
            
            digits.append(0)
        else:
            while digits and digits[-1] == 3:
                digits.pop()
            if digits:
                digits[-1] += 1
            if len(digits) == 1:
                digits.append(0)
                
    return val

ans = p037()

print(ans)
```
