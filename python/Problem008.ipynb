## Naive algorithm $\mathcal O (NM)$

Python just begs to be written as one-liner at the expense of computational efficiency.  A list of the products of all substrings of length $N$ is easily made, along with finding the maximum value.  This method is inefficient due to the fact that each element in the string of length $M$ has to be visited $N$ times, whereas the problem can be done with only going through the string once, regardless of $N$.

```python
%%time
from math import prod

def p008_naive(s: str, N: int) -> int:
    return max([prod(map(int,s[i:i+N])) for i in range(len(s)-N)])

s = "7316717653133062491922511967442657474235534919493496983520312774506326239578318016984801869478851843858615607891129494954595017379583319528532088055111254069874715852386305071569329096329522744304355766896648950445244523161731856403098711121722383113622298934233803081353362766142828064444866452387493035890729629049156044077239071381051585930796086670172427121883998797908792274921901699720888093776657273330010533678812202354218097512545405947522435258490771167055601360483958644670632441572215539753697817977846174064955149290862569321978468622482839722413756570560574902614079729686524145351004748216637048440319989000889524345065854122758866688116427171479924442928230863465674813919123162824586178664583591245665294765456828489128831426076900422421902267105562632111110937054421750694165896040807198403850962455444362981230987879927244284909188845801561660979191338754992005240636899125607176060588611646710940507754100225698315520005593572972571636269561882670428252483600823257530420752963450"
N = 13
ans = p008_naive(s, N)
print(ans)
```

## $\mathcal O (M)$

This is purely an algorithm problem and not a big math problem.  Here is how I do it only going through each digit once.  Start with the product variable `p` as $0$.  Then at each step, if `p` is $0$, then the next $N$ elements need to be multiplied together.  If a $0$ is encountered, then `p=0` and no more products are taken.  If instead `p != 0`, then the next digit is multiplied to `p` and the digit `N` elements back is divided.  The maximum seen `p` is then the answer.

The algorithm goes through the string once, so is of order $M$, the size of the string.

```python
%%time

def p008(s: str, N: int) -> int:
    pmax = 0
    imax = len(s)
    i = 0
    p = 0
    while i < imax:
        if p == 0:
            p = 1
            for j in range(N):
                if i == imax:
                    break

                p *= int(s[i])
                i += 1
                if p == 0:
                    break
        else:
            p //= int(s[i-N])
            p *= int(s[i])
            i += 1
        if p > pmax:
            pmax = p
            
            
    return pmax

s = "7316717653133062491922511967442657474235534919493496983520312774506326239578318016984801869478851843858615607891129494954595017379583319528532088055111254069874715852386305071569329096329522744304355766896648950445244523161731856403098711121722383113622298934233803081353362766142828064444866452387493035890729629049156044077239071381051585930796086670172427121883998797908792274921901699720888093776657273330010533678812202354218097512545405947522435258490771167055601360483958644670632441572215539753697817977846174064955149290862569321978468622482839722413756570560574902614079729686524145351004748216637048440319989000889524345065854122758866688116427171479924442928230863465674813919123162824586178664583591245665294765456828489128831426076900422421902267105562632111110937054421750694165896040807198403850962455444362981230987879927244284909188845801561660979191338754992005240636899125607176060588611646710940507754100225698315520005593572972571636269561882670428252483600823257530420752963450"
N = 13
ans = p008(s, N)
print(ans)
```
