# Problem 004:  Largest Palindrome Product

> A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is $9009 = 91 \times 99$.
>
>Find the largest palindrome made from the product of two $3$-digit numbers.


## $\mathcal O \left(10^{2N}\right)$

A straightforward approach to this problem is to search all products of $3$-digit numbers and find which are palindromes.  This results in an algorithm that will roughly be $\mathcal O \left(10^{2N}\right)$, but with $N=3$ it is feesible.

To test if a number is a palindrome, the digits of $x$ are reversed in a new variable $y$ and the two values are compared.

A few optimizations can be made for searching through all possible products of two $3$-digit numbers.  For example, $i*j = j*i$, so half the number of calculations are necessary by only considering $j > i$, for example.  Also, it is better to start from large values of $i$ and $j$ and then decrease them.  In this way the calculation can be stopped once it is impossible to find any value greater than a previously found palindrome.

Interestingly, with these modifications to the search the algorithm empiracally appears to scale as $\mathcal O \left(10^N\right)$.

The notes in Project Euler also point out that a $6$-digit pamildrome must be a multiple of $11$ so either $i$ or $j$ must also be a multiple of $11$.  I did not implement this below, but it would speed the search up a good bit.

```python
%%time
def is_palindrome(x: int) -> bool:
    xp = x
    y = 0
    while xp != 0:
        y = 10 * y + xp % 10
        xp //= 10
        
    return x == y

def p004(N: int) -> int:
    val = 0
    imax = pow(10,N)
    k = 0
    for i in range(imax - 1, imax//10 - 1, -1):
        if i * (imax - 1) < val:
            break
        for j in range(imax - 1, i - 1, -1):
            k = i * j
            if k > val and is_palindrome(k):
                #print(">  ", imax - i)
                val = k
            elif k < val:
                break
                
    return val


N = 3
ans = p004(N)
print(ans)
```
