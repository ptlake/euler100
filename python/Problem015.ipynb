# Problem 015:  Lattice Paths

> Starting in the top left corner of a $2 \times 2$ grid, and only being able to move to the right and down, there are exactly $6$ routes to the bottom right corner.
>
> > ![Problem 015 Image](../images/p015.png)
>
> How many such routes are there through a $20 \times 20$ grid?


## The math

This problem is more of a math problem than a coding problem; how can the number of ways be converted into something calculable?

Instead of the picture above, let's convert each path into a series of $r$ and $d$ for right and down, respectively:
> $(r, r, d, d) \qquad (r, d, r, d) \qquad (r, d, d, r)$
> 
> $(d, r, r, d) \qquad (d, r, d, r) \qquad (d, d, r, r)$

In looking at these it can be seen that there are always $4$ moves, $2$ of them $r$ and $2$ of them $d$.

To generalize this for a grid of size $N\times M$, there would always be $N+M$ moves, $N$ of them $r$ and $M$ of them $d$.  The problem becomes a matter of labeling $N$ of $N+M$ ordered objects, which is $N+M \choose N$.

## Python approach $\mathcal O(N+M)$

Knowing that all that needs to be done is to calculate a binomial coefficient, Python has a built in function `math.comb` for that.

From what I can see, the `math.comb` routine calculates the binomial coefficient explicitly as
$${n \choose m} = \frac{n!}{m!(n-m)!}.$$
Since the factorials are being calculated, the largest value will be the numerator and thus the time complexity is $\mathcal O(N+M)$

```python
%%time

from math import comb

def p015_python(N: int, M: int) -> int:
    return comb(N + M, N)

N = 20
M = 20
ans = p015_python(N, M)

print(ans)
```

## Write your own code approach $\mathcal O(N)$

Part of the point of this project is to over-engineer solutions.  So, why not write your own binomial coefficient algorithm?  There are so many ways of approaching this, and I present one below.

One approach is to use a recursive definition of a binomial coefficient:

$${n \choose m} = \frac{n - m + 1}{m}{n \choose m-1};\quad {n \choose 0} = 1$$

Better yet, a `reduce` call can avoid having to even write a loop over the values of $m$ in the above equation.

Since the binomial coefficients are symmetric between $N + M \choose N$ and $N + M \choose M$, taking the minimum of the two for the sum is more efficient in the case they are not equal.  The time comeplexity is $\mathcal O(N)$

An added benefit of this approach is that it minimizes the need to deal with excessively large values.  For example, with $N = M = 20$, $40!$ is $48$ digits long, whereas this algorithm will not have to deal with a number greater than $13$ digits long.

```python
%%time

from functools import reduce

def p015(N: int, M: int) -> int:
    N = min(N, M)
    a = N + M
    return reduce(lambda x, y: (x * (a - y)) // (y+1), range(N), 1)

N = 20
M = 20
ans = p015(N, M)

print(ans)
```
