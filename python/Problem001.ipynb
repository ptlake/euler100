# Problem 001: Multiples of 3 or 5

> If we list all the natural numbers below $10$ that are multiples of $3$ or $5$, we get $3$, $5$, $6$, and $9$. The sum of these multiples is $23$.
> 
> Find the sum of all the multiples of $3$ or $5$ below $1000$.


## Naive approach ($\mathcal O (NM)$)

With only going up to $1000$, it is easy enough to just go through all the values between $0$ and $1000$ (exclusive) and check if they are divisible by $3$ or $5$.

The resulting code is thus $\mathcal O (NM)$, where $N$ is the range of values tested ($N=1000$) and $M$ is the number of checks on divisibility ($M=2$, for the values $3$ and $5$).

```python
%%time
def p001_naive(N: int, div: list[int]) -> int:
    return sum([i for i in range(N) if any(i % j == 0 for j in div)])

N = 1_000
div = [3, 5]

ans = p001_naive(N, div)
print(ans)
```

## Better choice for large $N$ and small $M$ ($\mathcal O (2^M))$

A more direct way to calculate the value is to introduce triangle numbers, LCM (least common multiple), and the PIE (exclusion-inclusion principle).  In this case, the multiples of a value $x$ can be written as $kx$, for any natural number $k$.  Their sum up to the last multiple below $N$ is

$$ \mbox{sum of multiples of $x$ below $N$} = x \sum_{i=1}^{\lfloor\frac{N-1}{x}\rfloor} i.$$

Here, the $\lfloor x\rfloor$ function is the floor function, or integer division function, and the sum over integers are the Triangle Numbers.

Next, it is not enough to just find the sum over all the individual values of which we want multiples.  This would overcount the values that are multiples of 3 and 5, i.e. - multiples of 15.  In this fashion, the PIE (Inclusion--Exclusion Principle) is needed to find the multiples of any value on a list.  This especially is needed if more than two factors were to be considered.

Furthermore, the multiples of $x$ and $y$ are in general not just the product of $x$ and $y$, but rather the LCM (Least Common Multiple).  In the case asked here, the two factors are prime numbers so this complication arises only when trying to generalize to any list of factors.

This approach removes dependency on $N$ for the time complexity since the sum is evaluated in a single equation.  It does increase the time complexity in $M$, since the sum used in the PIE is $\mathcal O (2^{M})$.  Thus the previous algorithm could potentially be more effecient if there are a lot of factors being considered.


```python
%%time 
from itertools import combinations
from math import lcm, prod

def triangle_number(n: int) -> int:
    return (n * (n + 1)) // 2

def p001(N: int, div: list[int]) -> int:
    val = 0
    # PIE over all combinations of factors.
    parity = 1
    for i in range(1, len(div) + 1):
        for x in combinations(div, i):
            dnow = lcm(*x)
            val += parity * dnow * triangle_number((N - 1) // dnow)
        parity = -parity
        
    return val

N = 1_000
div = [3, 5]

ans = p001(N, div)
print(ans)
```

### Triangle Numbers
[OEIS: https://oeis.org/A000217](https://oeis.org/A000217)

[Wiki: https://en.wikipedia.org/wiki/Triangular_number](https://en.wikipedia.org/wiki/Triangular_number)

The $n^{th}$ triangle number, $a_n$, is the sum of values between $1$ and $n$.  The values of this sequence can be straightforwardly calculated using the equation 

$$a_n = \sum_{i=1}^{n} i = \frac{n(n+1)}{2}.$$
