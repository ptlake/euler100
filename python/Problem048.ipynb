# Problem 048:  Self Powers

> The series, $1^1 + 2^2 + 3^3 + \cdots + 10^{10} = 10405071317$.
>
> Find the last ten digits of the series, $1^1 + 2^2 + 3^3 + \cdots + 1000^{1000}$.

## Solution $\mathcal O\left(N \ln N\right)$

A one liner will do, and will be the most efficient way of doing it.  I could write out the loop and then do the binary exponentiation by hand, but it would be exactly the same as what is done in the one liner.

The sum is over all $N$ values, and each $n$ takes $\ln n$ steps for the exponentiation.  This makes the overall algorithm $\mathcal O(N\ln N)$.

```python
%%time

def p048(N: int, mod: int) -> int:
    return sum(pow(a, a, mod) for a in range(1, N+1)) % mod

N = 1_000
mod = pow(10,10)
ans = p048(N, mod)

print(ans)
```
