# Problem 022:  Names Scores

> Using names.txt (right click and 'Save Link/Target As...'), a 46K text file containing over five-thousand first names, begin by sorting it into alphabetical order. Then working out the alphabetical value for each name, multiply this value by its alphabetical position in the list to obtain a name score.
>
> For example, when the list is sorted into alphabetical order, COLIN, which is worth $3 + 15 + 12 + 9 + 14 = 53$, is the $938$th name in the list. So, COLIN would obtain a score of $938 \times 53 = 49714$.
>
> What is the total of all the name scores in the file?


## Solution

After processing the input file, it is just a one-liner.

First, the input file is a single line consisting of names in double quotes separated by commas.  Splitting on the string "\",\"", the resulting list has the names with the exception on the first and last on the list which will have a trailing double quote.  Conveniently, the names consist only of capital letters and no further processing is needed.

Next, `enumerate(sorted(names))` will produce a list of tuples containing the value of the alphabetical position, $0$ indexed, with the corresponding name in order.  The name score is found by using the `ord` function, which gives the alphabetical value of a letter shifted by $64$.

```python
%%time

def p022(names: list[str]) -> int:
    return sum((i+1) * (sum(map(ord, nnow)) - 64 * len(nnow)) for i, nnow in enumerate(sorted(names)))
        
with open("../data/p022_names.txt", 'r') as f:
    names = f.read().split("\",\"")
names[0] = names[0][1:]
names[-1] = names[-1][:-1]

ans = p022(names)

print(ans)
```
