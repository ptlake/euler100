# Problem 017:  Number Letter Counts

> If the numbers $1$ to $5$ are written out in words: one, two, three, four, five, then there are $3 + 3 + 5 + 4 + 4 = 19$ letters used in total.
>
> If all the numbers from $1$ to $1000$ (one thousand) inclusive were written out in words, how many letters would be used?
> 
> NOTE: Do not count spaces or hyphens. For example, $342$ (three hundred and forty-two) contains $23$ letters and $115$ (one hundred and fifteen) contains $20$ letters. The use of "and" when writing out numbers is in compliance with British usage.


## Hard coded value of $1000$

As a first pass and to answer the problem, I wrote a version that calculates the value specifically for the input of $1000$.  I consider every word that is used to construct the numbers and counted the number of times they are used.

> "one", "two", "three", "four", "five", "six", "seven", "eight", "nine" - These words appear in the one's position nine times every count of one hundred, i.e. - 90 times total.  Also, they appear in the hundred's position $100$ times, a piece.  Don't forget the extra "one" for "one thousand".
> 
> "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen", "seventeen", "eighteen", nineteen" - These "compound number" words appear once every $100$ times, a piece.
> 
> "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety" - These words appear ten times every $100$, i.e. - $100$ times, a piece.
> 
> "hundred" - This word appears in every number except the first $99$ and last $1$ - i.e. $900$ times.
> 
> "thousand" - Once in the last number.
> 
> "and" - This word appears $99$ times every $100$, except for the first $100$ - i.e. 891 times.

At this point, a program really isn't necessary and you can do it out by hand / with a calculator.  Regardless, here is a code with the above information.

```python
%%time

def p017_direct() -> int:
    val = 0
    # one
    val += len("one") * (9 * 10 + 100 + 1)
    # two
    val += len("two") * (9 * 10 + 100)
    # three
    val += len("three") * (9 * 10 + 100)
    # four
    val += len("four") * (9 * 10 + 100)
    # five
    val += len("five") * (9 * 10 + 100)
    # six
    val += len("six") * (9 * 10 + 100)
    # seven
    val += len("seven") * (9 * 10 + 100)
    # eight
    val += len("eight") * (9 * 10 + 100)
    # nine
    val += len("nine") * (9 * 10 + 100)
    # ten
    val += len("ten") * (10)
    # eleven
    val += len("eleven") * (10)
    # twelve
    val += len("twelve") * (10)
    # thirteen
    val += len("thirteen") * (10)
    # fourteen
    val += len("fourteen") * (10)
    # fifteen
    val += len("fifteen") * (10)
    # sixteen
    val += len("sixteen") * (10)
    # seventeen
    val += len("seventeen") * (10)
    # eighteen
    val += len("eighteen") * (10)
    # nineteen
    val += len("nineteen") * (10)
    # twenty
    val += len("twenty") * (10 * 10)
    # thirty
    val += len("thirty") * (10 * 10)
    # forty
    val += len("forty") * (10 * 10)
    # fifty
    val += len("fifty") * (10 * 10)
    # sixty
    val += len("sixty") * (10 * 10)
    # seventy
    val += len("seventy") * (10 * 10)
    # eighty
    val += len("eighty") * (10 * 10)
    # ninety
    val += len("ninety") * (10 * 10)
    # hundred
    val += len("hundred") * (100 * 9)
    # thousand
    val += len("thousand")
    # and
    val += len("and") * (99 * 9)
    return val

ans = p017_direct()

print(ans)
```

## `int` to `str` algorithm.

Despite not being an efficient way of calculating the desired value, consider writing an algorithm that takes an `int` of less than $1,000,000,000$ and writing it out into words.  Then the problem can be solved by adding the length of all the generated `str`'s.

```python
%%time

teen = {0: "ten",
        1: "eleven",
        2: "twelve",
        3: "thirteen", 
        4: "fourteen",
        5: "fifteen",
        6: "sixteen",
        7: "seventeen",
        8: "eighteen",
        9: "nineteen"}

ten = {0: "",
       1: "",
       2: "twenty",
       3: "thirty", 
       4: "forty",
       5: "fifty",
       6: "sixty",
       7: "seventy",
       8: "eighty",
       9: "ninety"}

one = {0: "",
       1: "one",
       2: "two",
       3: "three", 
       4: "four",
       5: "five",
       6: "six",
       7: "seven",
       8: "eight",
       9: "nine"}

def int_to_str(N: int) -> str:
    # handle 0 separately
    if N == 0:
        return "zero"
    words = []
    # first word will be "negative" for N < 0
    if N < 0:
        N = -N
        words.append("negative")
    # return empty string for numbers of a billion or more
    if N > 999_999_999:
        return ""

    # millions
    if N > 999_999:
        words.append(int_to_str(N // 1_000_000))
        words.append("million")

    # thousands
    if N % 1_000_000 > 999:
        words.append(int_to_str((N % 1_000_000) // 1_000))
        words.append("thousand")
    
    # hundreds
    if N % 1_000 > 99:
        words.append(int_to_str((N % 1_000) // 100))
        words.append("hundred")
    
    # and
    if N > 99 and N % 100:
        words.append("and")

    # 2-digit numbers
    tens = (N % 100) // 10
    ones = N % 10
    if tens == 1:
        words.append(teen[ones])
    else:
        if tens:
            words.append(ten[tens])
        if ones:
            words.append(one[ones])
        
    return "".join(words)

def p017_try2(N: int) -> int:
    return sum(len(int_to_str(n)) for n in range(1,N+1))

N = 1_000
ans = p017_try2(N)

print(ans)
```

## Counting each word occurence directly

The `int_to_str` method above is a straightforward $\mathcal O(N)$ approach, since each number is written out separately.  Instead, it is a $\mathcal O(1)$ algorithm to count the occurence of a word in all the numbers up to $N$, as is done in the hardcoded `p017_direct` routine.

```python
%%time

teen = {0: "ten",
        1: "eleven",
        2: "twelve",
        3: "thirteen", 
        4: "fourteen",
        5: "fifteen",
        6: "sixteen",
        7: "seventeen",
        8: "eighteen",
        9: "nineteen"}

ten = {0: "",
       1: "",
       2: "twenty",
       3: "thirty", 
       4: "forty",
       5: "fifty",
       6: "sixty",
       7: "seventy",
       8: "eighty",
       9: "ninety"}

one = {0: "",
       1: "one",
       2: "two",
       3: "three", 
       4: "four",
       5: "five",
       6: "six",
       7: "seven",
       8: "eight",
       9: "nine"}

def p017(N: int) -> int:
    val = 0
    M1 = N // 10
    M2 = N // 100
    M3 = N // 1_000
    M4 = N // 10_000
    M5 = N // 100_000

    for i in range(1,10):
        lnum = len(one[i])
        val += lnum * ((N + 10 - i) // 10 - (N + 90 - i) // 100)
        val += lnum * (100 * ((M2 + 9 - i) // 10) + (M2 % 10 == i) * (N % 100 + 1))
        val += lnum * (1_000 * ((M3 + 9 - i) // 10 - (M3 + 89 - i) // 100) + (M3 % 10 == i and M4 % 10 != 1) * (N % 1_000 + 1))
        val += lnum * (100_000 * ((M5 + 9 - i) // 10) + (M5 % 10 == i) * (N % 100_000 + 1))
    
    for i in range(10):
        lnum = len(teen[i])
        val += lnum * (((N + 90 - i) // 100))
        val += lnum * (1_000 * ((M3 + 89 - i) // 100) + (M4 % 10 == 1 and M3 % 10 == i) * (N % 1_000 + 1))
    
    for i in range(2,10):
        lnum = len(ten[i])
        val += lnum * (10 * ((M1 + 9 - i) // 10) + (M1 % 10 == i) * (N % 10 + 1))
        val += lnum * (10_000 * ((M4 + 9 - i) // 10) + (M4 % 10 == i) * (N % 10_000 + 1))
        
    # hundred
    lnum = len("hundred")
    val += lnum * (100 * (M2 - (M2 + 9) // 10) + (M2 % 10 != 0) * (N % 100 + 1))
    val += lnum * (100_000 * (M5 - (M5 + 9) // 10) + (M5 % 10 != 0) * (N % 100_000 + 1))

    # thousand
    lnum = len("thousand")
    val += lnum * (1_000 * (M3 - (M3 + 999) // 1_000) + (M3 > 0) * (N % 1_000 + 1))
    
    # and
    lnum = len("and")
    if N < 100:
        val -= lnum * N
    else:
        val -= lnum * 99
    val += lnum * (N - M2)
    if M5 > 0:
        if M3-1 < 100:
            val -= lnum * (M3-1) * 1_000
        else:
            val -= lnum * 99_000
        val += lnum * (1_000 * (M3 - 1 - (M3-1) // 100) + (M3 % 100 != 0) * (N % 1_000 + 1))

    return val

N = 1_000

ans = p017(N)
print(ans)

```
