# Problem 028:  Number Spiral Diagonals

> Starting with the number $1$ and moving to the right in a clockwise direction a $5$ by $5$ spiral is formed as follows:
>
> ***`21`***`22 23 24`***`25`*** <br>
> `20`***` 7`***` 8`***` 9`***`10` <br>
> `19  6`***` 1`***` 2 11` <br>
> `18`***` 5`***` 4`***` 3`***`12` <br>
> ***`17`***`16 15 14`***`13`*** <br>
>
> It can be verified that the sum of the numbers on the diagonals is $101$.
>
> What is the sum of the numbers on the diagonals in a $1001$ by $1001$ spiral formed in the same way?

## Naive solution $\mathcal O\left(N\right)$

Without too much ado, the numbers can be written out one by one and added together.  To find the pattern, notice that increment between successive elements of the diagonal is $\{2,2,2,2,4,4,4,4\}$ in the above $5\times 5$ square.  This goes on as $\{2,2,2,2,4,4,4,4,6,6,6,6,8,8,8,8,...\}$.

To code this, start with $a_0=1$ and $\delta_a = 2$.  Then $a_n = a_{n-1} + \delta_a$, where $\delta_a$ is incremented by two every four steps.  If you have a $N \times N$ square, then $\delta_a$ is incremented $\lfloor N/2\rfloor$ times, where $N$ must be an odd number to ensure that $1$ is in the center.  The answer is accumulated as a sum of $a_n$.

How about an even sided square?  That is easily accomidated in this scheme by changing $a_0 = 0$ and $\delta_a = 1$.  That is all that needs to be changed to get the answer.

This solution is $\mathcal O(N)$ for going through each value one by one.

```python
%%time

def p028_naive(N: int) -> int:
    if N % 2:
        val = 1
        da = 2
        a = 1
    else:
        val = 0
        da = 1
        a = 0
    for i in range(N//2):
        for j in range(4):
            a += da
            val += a
        da += 2
    return val

N = 1_000
ans = p028_naive(N)

print(ans)
```

## Direct calculation $\mathcal O\left(1\right)$

A closed form solution can be written for this problem as well, making a solution with constant time complexity.

Observe that the diagonal going up and to the right from the central $1$ is going to be perfect squares.  In fact, each diagonal will follow a quadtratic sequence.  To see this, consider $a_{n,m}$ to be the $m^\mbox{th}$ element along diagonal $n$, where the diagonals are indexed starting with the one going up and to the right and going counter-clockwise.  The recurrence relation is then found:
$$a_{n,m} = a_{n,m-1} + 8m - 2n$$
where $a_{n,0} = 1$ for all four diagonals.

To derive the quadratic sequence for each $a_{n,m}$, observe the recurrence relation for an arbitrary quadratic sequence $b_m$:
$$b_m = \alpha m^2 + \beta m + \gamma \implies b_m - b_{m-1} = \beta - \alpha + 2\alpha m \quad \mbox{and} \quad \gamma = b_0.$$
Then the correspondence between these two equations for an arbitrary $b_m$ and $a_{n,m}$ leads to:
$$a_{n,m} = 4m^2 + (4 - 2 n) m + 1.$$
The desired value $A_N$ is going to be the sum of all the diagonals up to $\lfloor N/2\rfloor$, making sure that the center value is only taken once:
$$\begin{align}
A_N & = 1 + \sum_{n=0}^3 \sum_{m=1}^{\lfloor N/2\rfloor} a_{n,m}\\
& = 1 + \sum_{m=1}^{\lfloor N/2\rfloor} 16 m^2 + 4 m + 4\\
& = 1 + \frac{8}{3}\lfloor N/2\rfloor(\lfloor N/2\rfloor+1)(2\lfloor N/2\rfloor + 1) + 2\lfloor N/2\rfloor(\lfloor N/2\rfloor+1) + 4 \lfloor N/2\rfloor\\
& = \frac{1}{3}\left(16 \lfloor N/2\rfloor^3 + 30 \lfloor N/2\rfloor^2 + 26 \lfloor N/2\rfloor + 3\right),
\end{align}$$
where the sum of integers and sum of integer squares functions are used and then rearranged to get the final cubic relation.

The same treatment can be done for the even sided squares to find the relation:
$$A_n = \frac{2}{3}\lfloor N/2\rfloor\left(8 \lfloor N/2\rfloor^2 + 3 \lfloor N/2\rfloor + 4\right).$$



```python
%%time

def p028(N: int) -> int:
    n = N // 2
    if N % 2:
        return (3 + n * (26 + n * (30 + 16 * n))) // 3
    else:
        return (n * (8 + n * (6 + 16 * n))) // 3

N = 1_001
ans = p028(N)

print(ans)
```
