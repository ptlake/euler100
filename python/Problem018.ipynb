## Every Route $\mathcal O (2^N)$

Well, they say you can do it this way, so why not. Using `itertools` allows for an easy way to go through all routes and sum up the values.  This approach considers all the subsets of the numbers in `range(N-1)`.  If the number is in the subset, then the route goes to the right after that row, otherwise to the left.

This is going to be $\mathcal O(2^N)$, where $N$ is the number of rows, because there is a factor of $2$ for deciding to go left or right on every row.

```python
%%time

from itertools import combinations

def p018_naive(a: list[list[int]]) -> int:
    val = 0
    N = len(a)
    for i in range(N):
        for x in combinations(range(N-1),i):
            k = 0
            y = [0]+[k := k + (j in x) for j in range(N)]
            val = max(val, sum([a[j][y[j]] for j in range(N)]))
    return val

a = list(map(lambda x: list(map(int,x.split())),"""75
95 64
17 47 82
18 35 87 10
20 04 82 47 65
19 01 23 75 03 34
88 02 77 73 07 63 67
99 65 04 28 06 16 70 92
41 41 26 56 83 40 80 70 33
41 48 72 33 47 32 37 16 94 29
53 71 44 65 25 43 91 52 97 51 14
70 11 33 28 77 73 17 78 39 68 17 57
91 71 52 38 17 14 91 43 58 50 27 29 48
63 66 04 68 89 53 67 30 73 16 69 87 40 31
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23""".split("\n")))


ans = p018_naive(a)

print(ans)
```

## Traveling up the triangle $\mathcal O(N^2)$

A faster algorithm is to travel from the bottom of the triangle, up.  At each row you store the largest value you can create starting at a given position going down.  This largest value will equal the input value at that position plus the larger of the two generated values below.  Upon working all the way up the triangle, the one remaining value is the maximum value.

This algorithm requires doing $n$ calculations on the $n^\mbox{th}$ row for all $N$ rows, hence it has a time complexity of $\mathcal O(N^2)$.

```python
%%time

def p018(a: list[list[int]]) -> int:
    val = 0
    N = len(a)
    b = a[N-1]
    for i in range(N-2, -1, -1):
        b = [a[i][j] + max(b[j:j+2]) for j in range(i+1)]
    return b[0]

a = list(map(lambda x: list(map(int,x.split())),"""75
95 64
17 47 82
18 35 87 10
20 04 82 47 65
19 01 23 75 03 34
88 02 77 73 07 63 67
99 65 04 28 06 16 70 92
41 41 26 56 83 40 80 70 33
41 48 72 33 47 32 37 16 94 29
53 71 44 65 25 43 91 52 97 51 14
70 11 33 28 77 73 17 78 39 68 17 57
91 71 52 38 17 14 91 43 58 50 27 29 48
63 66 04 68 89 53 67 30 73 16 69 87 40 31
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23""".split("\n")))


ans = p018(a)

print(ans)
```
