# Problem 003:  Largest Prime Factor

> The prime factors of $13195$ are $5$, $7$, $13$, and $29$.
> 
> What is the largest prime factor of the number $600851475143$?


## Naive approach $\mathcal O (\sqrt N)$

A fast way to solve the problem is to realize that you do not actually need to find any prime numbers.  Instead, loop over values, test divide the input by the value.  If it is divisible, divide all factors out of the input.  When the input value is reduced to 1, the last factor must be prime (otherwise a value before it would have been divided out), and that last factor will be the largest.  The inefficiency of this code is found in having to do trial division of all the composite factors between the primes.

As is, the time complexity would be $\mathcal O(N)$ because of the case that the input is prime the loop would have to execute all the way to $N$.  A considerably more efficient approach is to realize that the smallest prime divisor must be less than or equal to $\sqrt N$ or be $N$ itself.  The loop then does not go all the way to $N=1$, but rather until $N < i^2$, where $i$ is the loop variable.  The only added detail is that the case the input is divisible by the square of the largest prime factor, then $N$ will be reduced to $1$ and the largest prime factor is the previous loop variable.  Otherwise the remaining value in $N$ is the largest prime factor.  This code has a time complexity of $\mathcal O(\sqrt N)$, the worst case scenario being a prime input.

```python
%%time
def p003_naive(N: int) -> int:
    i = 1
    while N > i**2:
        i += 1
        while N % i == 0:
            N //= i

    if N == 1:
        N = i
        
    return N


N = 600_851_475_143
ans = p003_naive(N)
print(ans)
```

## ~~"Faster"~~ Different approach $\mathcal O (N \ln \ln N)$

Fine, the time scaling is asymptotically worse, and it is not necessary, but why not use the Sieve of Eratosthenes to find the primes first?  This might be a case of everything looks like a nail when you have a hammer, but it works.

A general lesson of these first few Euler Project problems is that the simple code is not only easier to write, but is actually faster than trying anything fancy.  But as the problems progress, having these tools developed is useful and sometimes necessary.

```python
%%time

def sieve_of_eratosthenes(num: int) -> list[int]:
    """ Returns primes up to and including num using a sieve.
    """
    sieve = [True] * ((num-1) // 2)
    pmax = int(num**0.5)//2
    for p in range(pmax):
        if sieve[p]:
            k = 2 * p + 3
            l = ((num - 1) // 2 - k**2 // 2) // k + 1
            sieve[k**2//2-1::k] = [False] * l
    return [2] + [2*i+3 for i in range((num-1)//2) if sieve[i]]


def p003(N: int) -> int:
    primes = sieve_of_eratosthenes(int(N**0.5))

    i = -1
    pnow = 1
    while N > pnow**2 and i+1 < len(primes):
        i += 1
        pnow = primes[i]
        while N % pnow == 0:
            N //= pnow

    if N == 1:
        N = pnow
        
    return N


N = 600_851_475_143
ans = p003(N)

print(ans)
```

It is worth noting that this second code is faster for a large prime input, say 1_000_000_007.
