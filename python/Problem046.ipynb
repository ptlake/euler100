# Problem 046:  Goldbach's Other Conjecture

> It was proposed by Christian Goldbach that every odd composite number can be written as the sum of a prime and twice a square.
>
> $$\begin{align}
9 = 7 + 2 \times 1^2\\
15 = 7 + 2 \times 2^2\\
21 = 3 + 2 \times 3^2\\
25 = 7 + 2 \times 3^2\\
27 = 19 + 2 \times 2^2\\
33 = 31 + 2 \times 1^2
\end{align}$$
>
> It turns out that the conjecture was false.
>
> What is the smallest odd composite that cannot be written as the sum of a prime and twice a square?

## Solution $\mathcal O\left(M^{3/2}\right)$

For every odd $n$, start with $j=0$.  Determine whether $n-2j^2$ is prime.  If it is, increment $n$, otherwise increment $j$ by $1$.  If $j$ becomes large enough that $n-j^2$ is negative, then $n$ is the answer.  By starting with $j=0$ you test for $n$ being composite or prime.

The prime sieve is the efficient way of determining if $n-2j^2$ is prime or not, and the extendable version is used here since it is not known _a priori_ how large $n$ will be.

For time scaling, I put it as $\mathcal O\left(M^{3/2}\right)$, where $M$ is the maximum value of $n$ checked.  The logic is that for every value $n$ there are $\sqrt n$ values of $j$ that need to be checked.  In actuality, the algorithm is faster then this estimate because not all the values of $j$ need to be checked and most numbers only need a small number of checks before a counterexample is found.  Without proof, the scaling looks more like $\mathcal O(M\ln M)$ if the number of checks of $j$ are summed up for large values (up to $10^7$).  Regardless, the check is slower than the prime generation.

There are apprarently only two known odd compoisite values that are not a sum of a primes number and twice a square, $5777$ and $5993$.  There are related to the Stern primes (Wikipedia: [Stern prime](https://en.wikipedia.org/wiki/Stern_prime)).  To avoid a (potentially/practically) infinite running code, and maximum value for $n$ is implemented in the code below.

```python
%%time

def extend_sieve(pow2: int, primes: list[int]) -> list[bool]:
    s = [True] * pow2
    # seive previous primes
    for p in primes:
        if p * p > 4 * pow2:
            break
        i0 = ((2 * pow2) // p + 1) * p
        if i0 % 2 == 0:
            i0 += p
        i0 = (i0 - 2 * pow2) // 2
        s[i0::p] = [False for _ in s[i0::p]]
    # search for new primes
    for i in range(pow2):
        if s[i]:
            primes.append(2*(pow2+i)+1)
    return s
    
def p046(N: int) -> int:
    pow2 = 1
    primes = []
    s = []
    
    cnt = 0
    n = 3
    ip = 0
    while True:
        if n > 2 * pow2:
            s += extend_sieve(pow2, primes)
            pow2 *= 2
        j = 0
        m = n
        while m > 1:
            if s[m // 2 - 1]:
                break
            m -= 4 * j + 2
            j += 1
        else:
            cnt += 1
            if cnt == N:
                return n
        n += 2
        if n > 1_000_000:
            print(f"Only found {cnt} terms.")
            break

    
    return 0

N = 1
ans = p046(N)

print(ans)
```
