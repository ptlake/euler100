# Problem 016:  Power Digit Sum

> $2^{15} = 32768$ and the sum of its digits is 3 + 2 + 7 + 6 + 8 = 26.
>
> What is the sum of the digits of the number $2^{1000}$?

## Python approach

In python, there is no challenge in this problem.  Calculate the $302$-digit number directly, sum the digits.

```python
%%time

def p016_python(N: int) -> int:
    return sum(map(int, str(pow(2, N))))

N = 1000
ans = p016_python(N)

print(ans)
```

## Write your own code approach

In many cases, I add this secondary approach because of some benefit over the more direct and/or naive implimentation.  Not here, there is no reason to do it this way.  This approach requires handling the data in a different way, but there is no benefit.  Not in time (it is $\mathcal O (N)$ as opposed to $\mathcal O (\ln N)$ with `pow`'s binary exponentiation), not in space (a $302$-digit number takes a chunk of space, but less than an array of that size).

Regardless, a list is created to store each digit of the number.  It is a lot easier to double a number than to write code to manually multiply two values, so the code does not use binary exponentiation.  The number of digits needed is calculated up front in order to initiate the array and not have to append to it dynamically.

```python
%%time

from math import log

def p016_manual(N: int) -> int:
    ndig_max = int(N * log(2.) / log(10.)) + 1
    ndig = 1
    digits = [None] * ndig_max
    digits[0] = 1
    
    for i in range(N):
        carry = 0
        for j in range(ndig):
            carry += 2 * digits[j]
            digits[j] = carry % 10
            carry = carry // 10
        while carry:
            digits[ndig] = carry % 10
            ndig += 1
            carry //= 10

    return sum(digits)

N = 1000
ans = p016_manual(N)

print(ans)
```
