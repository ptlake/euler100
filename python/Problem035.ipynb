# Problem 035:  Circular Primes

> The number, $197$, is called a circular prime because all rotations of the digits:  $197$, $971$, and $719$, are themselves prime.
>
> There are thirteen such primes below $100$: $2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79$, and $97$.
>
> How many circular primes are there below one million?

## Solution $\mathcal O (N \ln\ln N)$

There is not that much math necessary that has not been covered yet; the most technical part is the prime generation.  In this case, both the sieve itself and the prime numbers are useful.  Namely, all the primes up to $N$ are found, then each one is determined to be circular or not, adding $1$ to the answer each time one is found.

To determine is a prime is circular I use three tests.  One, if the prime is a single digit, it is circular.  Two, if there is an even number in any position of the prime (excepting $2$, itself), it is not circular.  Otherwise, all rotations of the number are tested.  A `for else` loop is used for the check, and the rotation is done mathematically as opposed to using strings.

Notice how an $n$-digit circular prime will be checked $n$ times in this method.  The number of circular primes is small s.t. the extra number of times the action is done is small in comparison to the number of primes there are.  The sieve value could be flipped for each check, but this adds some complications that are easier to just avoid.

Also, in the case that $N$ is not a power of $10$, the prime generation needs to go to the next full power of $10$ since the rotation of numbers below $N$ can be larger than $N$, in general.

The time scaling is $\mathcal O(N \ln \ln N)$, due to the prime generation.  For the search after generating the primes, there are $~N / \ln N$ primes, and $~ln N$ rotations per prime, making the overall scaling being faster than the prime generation, itself.

```python
%%time

def sieve_of_eratosthenes(num: int) -> list[int]:
    """ Returns primes up to and including num using a sieve.
    """
    sieve = [True] * ((num-1) // 2)
    pmax = int(num**0.5)//2
    for p in range(pmax):
        if sieve[p]:
            k = 2 * p + 3
            l = ((num - 1) // 2 - k**2 // 2) // k + 1
            sieve[k**2//2-1::k] = [False] * l
    return sieve, [2] + [2*i+3 for i in range((num-1)//2) if sieve[i]]
    
def p035(N: int) -> int:

    a = 1
    pow10 = [1] + [a := 10 * a for i in range(1, len(str(N)))]

    s, primes = sieve_of_eratosthenes(pow10[-1])
    evens = ('0', '2', '4', '6', '8')
    
    val = 0
    for p in primes:
        str_p = str(p)
        ndig = len(str_p)
        if ndig == 1:
            val += 1
            continue
        if any(e in str_p for e in evens):
            continue
        for i in range(1, ndig):
            p = (p % 10) * pow10[ndig-1] + p // 10
            if not s[p//2-1]:
                break
        else:
            val += 1
        
    return val
    
N = 1_000_000
ans = p035(N)
print(ans)
```
