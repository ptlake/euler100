# Problem 010:  Summation of Primes

> The sum of the primes below $10$ is $2 + 3 + 5 + 7 = 17$.
>
> Find the sum of all the primes below two million.


## Sieve algorithm $\mathcal O (N\ln\ln N)$

At this point, I have already demonstrated how the sieve of Eratosthenes to find primes and how it is suprerior to something like trial division.  See Problem 003, for example.  For that reason, I will just post this one solution using the sieve, where all that needs to be done is to sum the elements of the array of primes.  One benefit of doing a series of these problems is that the same concepts come up again and again.  Spending time writing efficient code for one problem can lead to 

The time scaling is for the sieve.

```python
%%time

def sieve_of_eratosthenes(num: int) -> list[int]:
    """ Returns primes up to and including num using a sieve.
    """
    sieve = [True] * ((num-1) // 2)
    pmax = int(num**0.5)//2
    for p in range(pmax):
        if sieve[p]:
            k = 2 * p + 3
            l = ((num - 1) // 2 - k**2 // 2) // k + 1
            sieve[k**2//2-1::k] = [False] * l
    return [2] + [2*i+3 for i in range((num-1)//2) if sieve[i]]


def p010(N: int) -> int:
    return sum(sieve_of_eratosthenes(N-1))


N = 2_000_000
ans = p010(N)

print(ans)
```
