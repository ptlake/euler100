# Problem 047:  Distinct Primes Factors

> The first two consecutive numbers to have two distinct prime factors are:
>
>  $$\begin{align}
14 &= 2 \times 7\\
15 &= 3 \times 5.
\end{align}$$
>
> The first three consecutive numbers to have three distinct prime factors are:
>
> $$\begin{align}
644 &= 2^2 \times 7 \times 23\\
645 &= 3 \times 5 \times 43\\
646 &= 2 \times 17 \times 19.
\end{align}$$
>
> Find the first four consecutive integers to have four distinct prime factors each. What is the first of these numbers?


## Solution $\mathcal O\left(M \ln \ln M\right)$

At first, it would appear that you need to factor every number up to the solution.  In effect, you do, but it can be done simultaneously with the prime sieve with some modifications.  First, for the sake of indexing and going through the list, all numbers are included in the sieve, not just odd numbers.  Next, you have to modify the sieve starting at the prime, not the prime squared.  Most versions of the sieve only need to identify the smallest prime factor of a number, of which the prime $p$ will never be for any value less than $p^2$.  Here, we need to account for all prime factors, so the sieve needs to be modified starting at $p$ itself.  When sieving, $1$ is added to each factor of $p$ to account for that prime being in the factorization.

After storing the number of unique prime factors in the sieve, a search is done to find $N$ prime factors in $N$ consecutive values.  These two processes, seiving and searching for the $N$ consecutive values, could be done simultaneously to be more efficient, but I have opted for them to be done separately for the simplicity in the code.  Since there is no way of knowing the size of the answer, I use the extendable sieve.

The search is an order $M$ process, where $M$ is the value that needs to be searched up to.  That makes the sieve the slowest part at $\mathcal O(M \ln \ln M)$.

```python
%%time

def extend_sieve(pow2: int, primes: list[int]) -> list[int]:
    s = [0] * pow2
    # seive previous primes
    for p in primes:
        i0 = ((pow2-1) // p + 1) * p - pow2
        s[i0::p] = [a+1 for a in s[i0::p]]
    # search for new primes
    for i in range(pow2):
        if not s[i]:
            s[i] = 1
            primes.append(pow2+i)
    return s
    
def p047(N: int) -> int:
    if N == 0:
        return 1
    if N == 1:
        return 2
    pow2 = 2
    primes = []
    s = [0,0]

    n = 2
    cnt = 0
    while True:
        if n >= pow2:
            s += extend_sieve(pow2, primes)
            pow2 *= 2
        if s[n] == N:
            cnt += 1
            if cnt == N:
                return n - N + 1
        else:
            cnt = 0
        
        n += 1
        if n > 1_000_000:
            break
    
    return 0

N = 4
ans = p047(N)

print(ans)
```
