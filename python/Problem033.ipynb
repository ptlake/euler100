# Problem 033:  Digit Cancelling Fractions

> The fraction $49 / 98$ is a curious fraction, as an inexperienced mathematician in attempting to simplify it may incorrectly believe that $49/98 = 4/8$, which is correct, is obtained by cancelling the $9$s.
>
> We shall consider fractions like, $30/50 = 3/5$, to be trivial examples.
>
> There are exactly four non-trivial examples of this type of fraction, less than one in value, and containing two digits in the numerator and denominator.
>
> If the product of these four fractions is given in its lowest common terms, find the value of the denominator.

## Solution

The search space for this problem is really small that there are not that many ways of going about this one.  Also, any way to generalize the problem would be a little arbitrary, so I limited this one to just solving this specific problem.

I guess one way would be to loop over all possible pairs of $2$-digit numbers and check if there is a digit to cancel out, but that seems more difficult.

Instead, I loop through three digits `i`, `j`, and `k`.  I let `i` be the repeated digit in the numerator and denominator, `j` be the one's digit of a number and `k` is the ten's digit of the other.  There's no point in casting the numbers into floats, instead if $ij / ki = j / k$, then that is the same as saying $ij \times k = ki \times j$.  The smaller of the values of `j` and `k` is multiplied to the numerator, the larger to the denominator.  The answer is the value of the denominator divided by the GCD of the two values.

```python
%%time

from math import gcd

def p033() -> int:
    num = 1
    den = 1
    for i in range(1, 10):
        for j in range(1, 10):
            for k in range(1, 10):
                j1 = 10 * i + j
                k2 = 10 * k + i
                if j * k2 == j1 * k and j1 != k2:
                    if j > k:
                        j, k = k, j
                    num *= j
                    den *= k

    return den // gcd(num, den)

ans = p033()

print(ans)
```
