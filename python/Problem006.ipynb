# Problem 006:  Sum Square Difference

> The sum of the squares of the first ten natural numbers is,
> $$1^2 + 2^2 + ... + 10^2 = 385.$$
>
> The square of the sum of the first ten natural numbers is,
> $$(1 + 2 + ... + 10)^2 = 55^2 = 3025.$$
> Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is $3025 - 385 = 2640$.
>
> Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.


## Naive approach $\mathcal O (N)$

With only $100$ values, it doesn't take much to just sum up the values directly.

```python
%%time

def p006_naive(N: int) -> int:
    return sum(range(N+1))**2 - sum([i**2 for i in range(N+1)])


N = 100
ans = p006_naive(N)
print(ans)
```

## $\mathcal O (1)$

The triangle numbers provide a direct calculation of the sum of integers:
$$\sum_{n=1}^N n = \frac{N(N+1)}{2}.$$
Likewise, the equation for the sum of squares is:
$$\sum_{n=1}^N n^2 = \frac{N(N+1)(2N+1)}{6}.$$
A little pen and pencil arithmetic yields:
$$\left(\sum_{n=1}^N n\right)^2 - \sum_{n=1}^N n^2 = \frac{N(N+1)(N-1)(3N+2)}{12}.$$

```python
%%time

def p006(N: int) -> int:
    return (N * (N + 1) * (N - 1) * (3 * N + 2)) // 12

N = 100
ans = p006(N)
print(ans)
```

One way to derive the sum of squares is to consider binomial coefficients and the hockey-stick identity:
$$\begin{align}
\sum_{n=2}^N {n \choose 2} & = {N+1 \choose 3} \\
\sum_{n=1}^N \frac{n(n-1)}{2} & = \frac{N(N+1)(N-1)}{6} \\
\frac{1}{2}\sum_{n=1}^N n^2 - \frac{N(N+1)}{4} & = \frac{N(N+1)(N-1)}{6}\\
\implies \sum_{n=1}^N n^2 & = \frac{N(N+1)(2N+1)}{6}
\end{align}$$
