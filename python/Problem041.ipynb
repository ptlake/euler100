# Problem 041:  Pandigital Prime

> We shall say that an $n$-digit number is pandigital if it makes use of all the digits $1$ to $n$ exactly once. For example, $2143$ is a $4$-digit pandigital and is also prime.
>
> What is the largest $n$-digit pandigital prime that exists?

## Code before thinking Solution

So, we just need to generate all the $n$-digit pandigital numbers, sort them in descending order, and test for primality until you get a hit.

Using `permutations` from `itertools` generates the palindromes.  To get them in numerical order, start at the largest length, $9$, and decrease the length.  For within a set of $n$ digits, sort the list in reverse order first, and `permutations` will produce the palindromes in descneding order.  I did take liberty and generalize the `perm_prime` function to take a list of digits in an arbitrary order, and also to avoid running the `is_prime` function for the case where there are repeating digits.  Also, if the leading didgit is a $0$, then the loop is exited early, since presumably there cannot be leading $0$'s in a number palindrome.

The `is_prime` function is just a brute trial division algorithm, askewing the need to generate primes.

Even though the code unnecessarily goes through the pandigital numbers of length $9$ and $8$, it still takes about a second to run on this machine.

```python
%%time

from itertools import permutations

def is_prime(n: int) -> bool:
    i = 2
    while i**2 <= n:
        if not n % i:
            return False
        i += 1
    return True

def perm_prime(x: list[int]) -> int:
    x.sort(reverse=True)
    yp = []
    for y in permutations(x):
        if y[0] == 0:
            break
        if y != yp:
            if is_prime(int(''.join(map(str, y)))):
                return int(''.join(map(str, y)))
            yp =  y
    return 0


def p041() -> int:
    for i in range(9,0,-1):
        val = perm_prime(list(range(i,0,-1)))
        if val:
            return val
    return 0
    
ans = p041()

print(ans)
```

## Think before you code Solution

Hindsight is 20/20.

A simple trick can be done to weed out certain palindromes.  Namely, if the sun of digits is divisible by $3$, then any palindrome will be divisible by $3$.  No need to run through all the permutations of that set of digits.  This is exactly the situation for$9$- and $8$-digit pandigital numbers.  Below has this one check in it before all the permutations are run.  Granted, similar checks such as checking there is an odd digit somewhere in the list could be implemented, but are irrelevant for the current code.

The question becomes, did you think of this possibility in less than the $1$ second it took to run the entire code without weeding out the $8$- and $9$-digit pandigital numbers?

```python
def perm_prime(x: list[int]) -> int:
    if not sum(x) % 3 and x != [3]:
        if x == [3]:
            return 3
        return 0
    x.sort(reverse=True)
    yp = []
    for y in permutations(x):
        if y[0] == 0:
            break
        if y != yp:
            if is_prime(int(''.join(map(str, y)))):
                return int(''.join(map(str, y)))
            yp =  y
    return 0
```
