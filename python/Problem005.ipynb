# Problem 005:  Smallest Multiple

> $2520$ is the smallest number that can be divided by each of the numbers from $1$ to $10$ without any remainder.
>
> What is the smallest positive number that is evenly divisible by all of the numbers from $1$ to $20$?

## Pythonic approach $\mathcal O \left(N \ln N\right)$

In Python, it is usually worth just using built in functions as opposed to writing out more sophisticated code.  Time scaling be damned - the speed of the built in functions not being written in Python.  In this case, the problem can be solved with a one-liner using LCM.  The smallest number divisible by all elements of the set is the same as the Least Common Multiple of the set.  I give the time scaling as $\mathcal O \left(N\ln N\right)$ because the LCM algorithm will presumably call a GCD algorithm, costing $\mathcal O (\ln N)$, $N$ times.

```python
%%time

from math import lcm

def p005_naive(N: int) -> int:
    return lcm(*range(1, N+1))


N = 20
ans = p005_naive(N)
print(ans)
```

## "Faster" Approach $\mathcal O \left(N \ln \ln N\right)$

A different approach can be taken to calculate the LCM of numbers below $N$.  Here, all the primes below $N$ are found, and will necessarily be a factor in the answer.  Further more, the power of a prime $p$ will be the largest $n$ s.t. $p^n < N$.  In the code below, I just keep the calculation using integer types by multiplying factors of $p$ together until the product is greater than $N$.  The value $n$ can also be more directly found by using logarithms, $n = \left \lfloor \frac{\ln N}{\ln p} \right\rfloor$, but requires using floating point math and all its complications.  As for time complexity, there are approximately $\frac{N}{\ln N}$ primes that need to be calculated, and ech one takes on the order of $\ln N$ calculations.  This $\mathcal (N)$ calculation is thus faster than the prime generator, whose complexity is quoted above.

```python
%%time

from math import sqrt

def sieve_of_eratosthenes(num: int) -> list[int]:
    """ Returns primes up to and including num using a sieve.
    """
    sieve = [True] * ((num-1) // 2)
    pmax = int(sqrt(num))//2
    for p in range(pmax):
        if sieve[p]:
            k = 2 * p + 3
            l = ((num - 1) // 2 - k**2 // 2) // k + 1
            sieve[k**2//2-1::k] = [False] * l
    return [2] + [2*i+3 for i in range((num-1)//2) if sieve[i]]


def p005(N: int) -> int:
    primes = sieve_of_eratosthenes(N)
    val = 1
    for p in primes:
        powp = p
        while p * powp <= N:
            powp *= p
        val *= powp
    return val

N = 20
ans = p005(N)
print(ans)
```

As a last note, these time complexities are a little vacuous since at some point it will be the speed of handling large integers that will really effect the timing.  With $N\simeq 20$, the code will run fast enough if the code is reasonable.
