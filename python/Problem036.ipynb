# Problem 036:  Double-base Palindromes

> The decimal number, $585 = 1001001001_2$ (binary), is palindromic in both bases.
>
> Find the sum of all numbers, less than one million, which are palindromic in base $10$ and base $2$.
>
> (Please note that the palindromic number, in either base, may not include leading zeros.)

## Naive Solution $\mathcal O (N \ln N)$

With only needing to check a million values, just loop through all of them and check if they are a palindrome in both the bases.

To check for a palindrome, the digits of the number in base $b$ are stored in a list, and the ckeck is if the list and the list reversed are the same.

With checking all $N$ values, needing to convert it to two given bases in $\ln n$ steps for a value $n$, makes the overall time scaling $\mathcal O(N\ln N)$.

```python
%%time

def is_palindrome(x: int, b: int) -> bool:
    x_list = []
    while x:
        x_list.append(x % b)
        x //= b
    return x_list == x_list[::-1]

def p036_naive(N: int, b1: int, b2: int) -> int:
    val = 0
    for i in range(1, N):
        if is_palindrome(i, b1) and is_palindrome(i, b2):
            val += i
    return val
    
N = 1_000_000
b1 = 10
b2 = 2
ans = p036_naive(N, b1, b2)
print(ans)
```

## Faster Solution $\mathcal O (\sqrt N \ln N)$

The number of palindromes is small compared to the number of values up to $N$.  A faster approach is to construct a palindrome in one base and then check it in the other.

Since a palindrome can be made by setting half the digits in a number, the number of palindromes less than $N$ is roughly $\sqrt N$.  Thus only $\sim\sqrt N$ checks need to be made, making this algorithm's time scaling $\mathcal O(\sqrt N \ln N)$

```python
%%time

def is_palindrome(x: int, b: int) -> bool:
    x_list = []
    while x:
        x_list.append(x % b)
        x //= b
    return x_list == x_list[::-1]
    
def palindrome_generator(b: int, nmax: int) -> int:
    left = 0
    powb = b
    nd_odd = True
    n = 0
    while n < nmax:
        n = left
        ip = n
        if nd_odd:
            ip //= b
        while ip:
            n = b * n + ip % b
            ip //= b

        left += 1
        if left == powb:
            if nd_odd:
                left //= b
            else:
                powb *= b
            nd_odd = not nd_odd

        yield n

def p036(N: int, b1: int, b2: int) -> int:
    val = 0

    for n in palindrome_generator(b1, N):
        if is_palindrome(n, b2):
            val += n
        
    return val
    
N = 1_000_000
b1 = 10
b2 = 2
ans = p036(N, b1, b2)
print(ans)
```
